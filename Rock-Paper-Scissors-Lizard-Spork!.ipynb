# Basic Rock-Paper-Scissors code

Let's start off with the basic code of Rock, Paper, Scissors code.
Since there about 9 possible combinations between user and computer.
This is fairly straight forward to build.

```python
#I broke the game into 3 parts and create the functions
# start()--> Prompts user if they would like to play the game
def start():
    x = input('Play the game? \n(yes/no)?: \n')
    if x.lower() =='yes':
        play()
            
    elif x.lower() == 'no':
        print('\nPlay next time!\n')
        
    else:
        print('Please choose either(yes/no)!') #This is to prevent early termination of game incase of typo
        start()

# first_part() --> choose rock, paper or scissors
def play():
    while True:
        user_choice = input('\nChoose (rock, paper or scissors): ')
        if user_choice.lower() in ['rock','paper','scissors']:
            choices = ['rock','paper','scissors']
            com_choice = random.choice(choices)
            return evaluate(com_choice, user_choice)
        else:
            print('Please choose only (rock, paper or scissors)!\n') #This is to prevent early termination of game in case of typo
            
            
def evaluate(com_choice, user_choice):
    if user_choice== com_choice:
        print('Both players selected', str(com_choice),'. It is a tie!')
        
    elif user_choice == 'rock':
        if com_choice == 'scissors':
            print('Rock smashes scissors! You win!')
        else:
            print('Paper covers rock! You lose.')
            
    elif user_choice == 'paper':
        if com_choice == 'rock':
            print('Paper covers rock! You win!')
        else:
            print('Scissors cuts paper! You lose.')
            
    elif user_choice == 'scissors':
        if com_choice == 'paper':
            print('Scissors cuts paper! You win!')
        else:
            print('Rock smashes scissors! You lose.')
    return play_again()
            
#allows user to choose to play again 
def play_again():
    x = input('Play the game again? \n(yes/no)?: ')
    if x.lower() =='yes':
        play()
    elif x.lower() == 'no':
        print('\nPlay again next time!')
    else:
        print('\nChoose either (yes/no)!')
        play_again()
    
```

# Rock-Paper-Scissors using Win-Lose Matrix

However, since we are going to build Big Bang Theory's Rock, Paper, Scissors, Lizard and Spork. It is not advisable to list out all possible combinations hence, we should use the Win-Lose Matrix to build the code instead.

```python
#Mapping between moves and numbers
game_chart = {0:"rock", 1:"paper", 2:"scissors", 3:"lizard", 4:"Spock"}
 
# Win-lose matrix for traditional game
rps_chart = [[-1, 1, 0], [1, -1, 2], [0, 2, -1]]

def rps_start():
    x = input('Play the game? \n(yes/no)?: \n')
    if x.lower() =='yes':
        rps_play()
    elif x.lower() == 'no':
        print('\nPlay next time!\n') 
    else:
        print('Please choose either(yes/no)!') #This is to prevent early termination of game incase of typo
        rps_start()

def rps_play():
    while True:
        rps_user_choice = input('\nChoose (rock, paper or scissors): ')
        if rps_user_choice.lower() in ['rock','paper','scissors']:
            rps_com_choice = random.randint(0, 2)
            
            if rps_user_choice.lower() == "rock":
                rps_user_move = 0
            elif rps_user_choice.lower() == "paper":
                rps_user_move = 1
            elif rps_user_choice.lower() == "scissors":
                rps_user_move = 2
                
            return evaluate(rps_com_choice, rps_user_move, rps_user_choice)
        else:
            print('Please choose only (rock, paper or scissors)!\n') #This is to prevent early termination of game in case of typo
            
def evaluate(rps_com_choice, rps_user_move, rps_user_choice):
    winner = rps_chart[rps_user_move][rps_com_choice]
    com = game_chart[rps_com_choice]
    print("Sheldon chooses", str(com),'!')
    if winner == rps_user_move:
        print(str(rps_user_choice), 'beats', str(com), '!')
        print("\nYou Win!!!\n")
    elif winner == rps_com_choice:
        print(str(com), 'beats', str(rps_user_choice), '!')
        print("\nSheldon Wins!!!\n")
    else:
        print("\nIts a Draw!!!\n")
        
    return rps_play_again()

def rps_play_again():
    x = input('Play the game again? \n(yes/no)?: ')
    if x.lower() =='yes':
        rps_play()
    elif x.lower()=='no':
        print('\nPlay again next time!\n')
        w = input('Turn off the device? \n(yes/no):')
        if w =='yes':
            print('\nGoodBye!')
            print('Game Device switched off.\n')
        elif w =='no':
            game_device()
    else:
        print('Choose(yes/no)!')
        rps_play_again()
```
